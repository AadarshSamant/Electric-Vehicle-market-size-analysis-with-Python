```python
#Task(2).you suspect----column.
import numpy as np

col="mean - Energy consumption [kWh/100 km]"

Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR=Q3-Q1

lower_bound= Q1-1.5 * IQR
upper_bound=Q3+1.5*IQR

outliners=df[(df[col]< lower_bound) | (df[col]> upper_bound)]

print("outliners in energy consumption:\n")
print(outliners[["Car full name","Make","Model",col]])
```

```python
##Task(4):Build an EV recommendation class.

import pandas as pd 

class EV_Recommender:
    """
    A Recommendation system for Electric Vehical (Evs).
    """

    def __init__(self,dataframe: pd.DataFrame):
        self.df = dataframe

    def recommend(
        self,
        budget: float,
        min_range: float,
        min_battery: float,
        sort_by: str = "Range (WLTP) [km]",
        top_n: int = 3
    ) -> pd.DataFrame:
        """
        Recommend Evs based on user criteria.
        """

        required_columns = [
            "Car full name", "Make", "Model",
            "Minimal price (gross) [PLN]",
            "Range (WLTP) [km]",
            "Battery capacity [kWh]"]
        for  col in required_columns:
            if col not in self.df.columns:
                raise KeyError(f"column '{col}' not found in dataset.")

        filtered=self.df[
            (self.df["Minimal price (gross) [PLN]"] <= budget) &
            (self.df["Range (WLTP) [km]"] >= min_range) &
            (self.df["Battery capacity [kWh]"] >=min_battery)
        ][required_columns]

        filtered = filtered.sort_values(by=sort_by, ascending=False)
        return filtered.head(top_n).reset_index(drop=True)
    

if __name__== "__main__":
    recommender = EV_Recommender(df)
    
try:
         budget = float(input("Enter your budget (PLN):"))
         min_range = float(input("Enter desired minimum range (km): "))
         min_battery = float(input("Enter minimum battery capacity (kWh)"))

         result= recommender.recommend(
             budget=budget,
             min_range=min_range,
             min_battery=min_battery,
             sort_by="Range (WLTP) [km]",
             top_n=3)

         if result.empty:
             print("\n No Evs found that match your criteria.")
         else:
             print("\n Top Recommended Evs:\n")
             print(result)

except ValueError:
    print("Please Enter numeric values for budget, range, and battery capacity.")
```

```python
#Task(5).Inferential Statistics– Hypothesis Testing 

import pandas as pd
from scipy.stats import ttest_ind

tesla_power = df[df["Make"] == "Tesla"]["Engine power [KM]"].dropna()
audi_power = df[df["Make"]== "Audi"]["Engine power [KM]"].dropna()

t_stat, p_value = ttest_ind(tesla_power, audi_power, equal_var=False)

print("Two-Sample T-Test: Tesla vs Audi Engine Power")
print("-----------------------------------------------")
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.5f}")

alpha = 0.05 
if p_value < alpha:
    print("\n Result: Reject the Null Hypothesis")
    print("There IS a significant difference in engine power between Tesla and Audi.")

else:
    print("\n Result: Fail to reject the Null hypothesis.")
    print("There is NO significant difference in Engine Power between Tesla and Audi.") 

```
